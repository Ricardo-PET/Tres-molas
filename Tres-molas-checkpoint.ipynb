O objetivo deste projeto é simular o movimento de uma massa ligada a três pontos por meio de molas. Ele foi escrito em Python e usará conceitos de Física computacional na resolução de EDOs, portanto requer um conhecimento prévio no assunto para a melhor compreensão [1].

**Importando as bibliotecas:**

```python
from matplotlib import pyplot as plt
import numpy as np
from matplotlib.pyplot import xlabel, ylabel, legend
```

Considere um sistema com um corpo no ponto $P$ de massa $m$ ligado, por molas, a três pontos $(L,0)$, $(0,L)$ e $(L,2L)$ em um plano $x-y$. Além disso, as três molas tem constante elástica $k$, e estão relaxadas se tiverem um comprimento $L$, como na figura abaixo:

<img src="mola1.1.jpg">

Para descrevermos o movimento do corpo no ponto $P$ usaremos a mecânica Lagrangiana. Primeiro devemos achar a Lagrangiana do sistema, ela é dada de forma genérica por:


$$\mathcal{L} = T - V$$

A energia cinética $T$ da partícula e sua velocidade ao quadrado $v^2$ em coordenadas cartesianas (x,y) são dadas respectivamente por:

$$T = \dfrac{mv^2}{2}$$

$$v^2 = \dot{x}^2 + \dot{y}^2$$

Logo, 

$$\boxed{T = \dfrac{m}{2}(\dot{x}^2 + \dot{y}^2)}$$

Como a única fonte de potencial no sistema são os potenciais elásticos das três molas, a energia potencial $V$ do sistema é dada pela soma das energias potências de cada mola. A energia potêncial elástica de uma mola ideal pode ser encontrada através da lei de Hook, e ela é dada por:

$$V_{elastica} = \dfrac{k\delta^2}{2}$$

Onde $\delta$ é a deformação da mola em relação ao seu estado relaxado. No caso das molas do sistema seu estado relaxado é quando o seu comprimento é $L$, logo, para encontrar o valor de $\delta_i$ para cada mola também precisamos encontrar o seu tamanho em cada instante de tempo. Considere a imagem seguinte: 

<img src="mola1.2.jpg">

Em um estado qualquer, considere que o tamanho das molas (cada uma ligada a seus respectivos pontos) são dados por $r_1$, $r_2$ e $r_3$. Por pitágoras, podemos encontrar estes comprimentos e relacioná-los ao $L$ para encontrarmos os valores dos $\delta_i$:

$$\delta_1 = r_1 -L = \sqrt{x^2 + (y-L)^2}-L$$

$$\delta_2 = r_2 -L = \sqrt{(x-L)^2 + (y-2L)^2}-L$$

$$\delta_3 = r_3 -L = \sqrt{(x-L)^2 + y^2}-L$$

Assim, a energia potencial elástica total do sistema é dada por: 

$$V = V_1 + V_2 + V_3$$

$$\boxed{V = \dfrac{k}{2} \left [\left (\sqrt{x^2 + (y-L)^2}-L \right)^2 + \left (\sqrt{(x-L)^2 + (y-2L)^2}-L \right)^2 + \left (\sqrt{(x-L)^2 + y^2}-L \right)^2   \right]}$$

Finalmente, a lagrangiana do sistema com três molas em coordenadas cartesianas é dada por:

$$\boxed{\mathcal{L} = \dfrac{m}{2}(\dot{x}^2 + \dot{y}^2) - \dfrac{k}{2} \left [\left (\sqrt{x^2 + (y-L)^2}-L \right)^2 + \left (\sqrt{(x-L)^2 + (y-2L)^2}-L \right)^2 + \left (\sqrt{(x-L)^2 + y^2}-L \right)^2   \right]}$$

Para encontrarmos as EDOs que descrevem o movimento da partícula em $P$ utilizaremos a equação de Euler. Esta equação para coordenadas genéricas $q_i$ é:

$$\dfrac{\partial \mathcal{L}}{\partial q_i}-\dfrac{d}{dt}\dfrac{\partial \mathcal{L}}{\partial \dot{q}_i} = 0$$

Resolvendo a equação de euler para $q_i = x$:

$$\dfrac{\partial \mathcal{L}}{\partial x} - \dfrac{d}{dt} \dfrac{\partial \mathcal{L}}{\partial \dot{x}} = 0$$

Onde, 

$$\dfrac{\partial \mathcal{L}}{\partial x} = -k \left [3x-2L-L \left ( \dfrac{x}{\sqrt{x^2 +(y-L)^2}} + \dfrac{(x-L)}{\sqrt{(x-L)^2 +(y-L)^2}} + \dfrac{x}{\sqrt{(x-L)^2 + y^2}}   \right ) \right ]$$

$$\dfrac{\partial \mathcal{L}}{\partial \dot{x}} = m \dot{x}$$

$$\dfrac{d}{dt} \dfrac{\partial \mathcal{L}}{\partial \dot{x}} = m \ddot{x}$$

Logo, 

$$\boxed{\ddot{x} =  -\dfrac{k}{m} \left [3x-2L-L \left ( \dfrac{x}{\sqrt{x^2 +(y-L)^2}} + \dfrac{(x-L)}{\sqrt{(x-L)^2 +(y-L)^2}} + \dfrac{x}{\sqrt{(x-L)^2 + y^2}}   \right ) \right ]}$$

Resolvendo a equação de euler para $q_i = y$:

$$\dfrac{\partial \mathcal{L}}{\partial y} - \dfrac{d}{dt} \dfrac{\partial \mathcal{L}}{\partial \dot{y}} = 0$$

Onde, 

$$\dfrac{\partial \mathcal{L}}{\partial y} = -k \left [3y-3L-L \left ( \dfrac{(y-L)}{\sqrt{x^2 +(y-L)^2}} + \dfrac{(y-2L)}{\sqrt{(x-L)^2 +(y-L)^2}} + \dfrac{y}{\sqrt{(x-L)^2 + y^2}}   \right ) \right ]$$

$$\dfrac{\partial \mathcal{L}}{\partial \dot{y}} = m \dot{y}$$

$$\dfrac{d}{dt} \dfrac{\partial \mathcal{L}}{\partial \dot{y}} = m \ddot{y}$$

Logo, 

$$\boxed{\ddot{y} = -\dfrac{k}{m} \left [3y-3L-L \left ( \dfrac{(y-L)}{\sqrt{x^2 +(y-L)^2}} + \dfrac{(y-2L)}{\sqrt{(x-L)^2 +(y-L)^2}} + \dfrac{y}{\sqrt{(x-L)^2 + y^2}}   \right ) \right ]}$$

**Criando uma função com as EDOs**

```python
def f(r,t):
    x,x_1p,y,y_1p = r[0],r[1],r[2],r[3]

    x_2p = -(k/m)*(3*x-2*L -L*(x/(x**2 +(y-L)**2)**(1/2) + (x-L)/((x-L)**2 +(y-2*L)**2)**(1/2) + (x-L)/((x-L)**2 +y**2)**(1/2)))
    y_2p = -(k/m)*(3*y-3*L -L*((y-L)/(x**2 +(y-L)**2)**(1/2) + (y-2*L)/((x-L)**2 +(y-2*L)**2)**(1/2) + y/((x-L)**2 +y**2)**(1/2)))
    
    return np.array([x_1p,x_2p,y_1p,y_2p])
```

**Criando uma função para o método Runge Kutta:**

O método Runge Kutta de quarta ordem é um método computacional bastante preciso, usado para resolver equações diferenciais. Na verdade, ele é um conjunto de métodos para resolver EDOs, que incluem por exemplo o método de Euler. Seu funcionamento basicamente consiste em receber dados iniciais de uma função e dizer onde ela estará no passo seguinte.

```python
def RK4(r,t):
    k1 = h*f(r,t)
    k2 = h*f(r+k1/2,t+h/2)
    k3 = h*f(r+k2/2,t+h/2)
    k4 = h*f(r+k3,t+h)
    return r + (1/6)*(k1+2*k2+2*k3+k4)
```

**Definindo os parâmetros e constantes**

A nossa simulação ocorrerá entre o de tempo $a$ e $b$, onde $h$ é o intervalo de tempo entre as $N$ iterações. Vamos considerar a massa $m$ da partícula, o comprimento $L$ e a constante $k$ como tendo todos o valor 1.0 (guardado as devidas unidades).

```python
a = 0
b = 20.0
L = 1.0
m = 1.0
k = 1.0
N = 500
h = (b-a)/N
```

**Exemplo 1**

Vamos considerar que a partícula se encontra inicialmente em repouso próximo ao ponto $(0,L)$ e simetricamente distante dos pontos $(L,0)$ e $(L,2L)$. As listas $X$ e $Y$ servem para guardar todas as posições da partículas no plano $x-y$, e o $r$ é usado para guardar as posições $x$ e $y$, bem como as velocidades $\dot{x}$ e $\dot{y}$ para cada iteração.

```python
x, x_1p, y, y_1p = 0.2*L,0,L,0
X = []
Y = []
r = np.array([x,x_1p,y,y_1p])
```

**Resolvendo as EDOs**

```python
t = a
while(t<b):
    X.append(r[0])
    Y.append(r[2])
    r = RK4(r,t) 
    t+=h
```

**PLotando os gráficos**

Podemos perceber que com esta configuração a partícula tem um movimento oscilatório uniforme apenas em $x$, justamente por causa da simetria das distâncias entre os pontos $(L,0)$, $(L,2L)$ e a partícula. OBS: O ponto vermelho é a posição inicial da partícula.

```python
plt.xlabel("x [m]")
plt.ylabel("y [m]")
plt.plot(L, 0, 'bo')
plt.plot(0, L, 'bo')
plt.plot(L, 2*L, 'bo')
plt.plot(X, Y, label='P', color = 'g')
plt.plot(X[0], Y[0],'ro')
plt.legend()
plt.xlim(-0.2,2.2*L)
plt.ylim(-0.2,2.2*L)
plt.show()

plt.ylabel("x [m]")
plt.plot(X, label='x')
plt.show()
plt.ylabel("y [m]")
plt.plot(Y, label='x')
plt.show()
```

**Exemplo 2**

Vamos considerar que a partícula se encontra novamente em repouso no começo, e perto do ponto $(L,2L)$. Nesta situação, no instante inicial a mola ligada ao ponto $(L,2L)$ irá repelir fortemente a partícula, enquanto as outras irão atrair, gerando um movimento caótico.

```python
x, x_1p, y, y_1p = L,0,1.8*L,0
X = []
Y = []
r = np.array([x,x_1p,y,y_1p])
```

**Resolvendo as EDOs e Plotando os gráficos**

```python
t = a
while(t<b):
    X.append(r[0])
    Y.append(r[2])
    r = RK4(r,t) 
    t+=h

plt.xlabel("x [m]")
plt.ylabel("y [m]")
plt.plot(L, 0, 'bo')
plt.plot(0, L, 'bo')
plt.plot(L, 2*L, 'bo')
plt.plot(X, Y, label='P', color = 'g')
plt.plot(X[0], Y[0],'ro')
plt.legend()
plt.xlim(-0.2,2.2*L)
plt.ylim(-0.2,2.2*L)
plt.show()

plt.ylabel("x [m]")
plt.plot(X, label='x')
plt.show()
plt.ylabel("y [m]")
plt.plot(Y, label='x')
plt.show()
```

**Exemplo 3**

Vamos considerar que a partícula inicialmente está no ponto $(L,L)$, mas que desta vez tem um motvimento inicial baixo. Por causa da pouca energia cinética ela não irá muito longe, mantendo seu movimento diminuto próximo ao ponto $(L,L)$.

```python
x, x_1p, y, y_1p = L,0.3,L,0.2
X = []
Y = []
r = np.array([x,x_1p,y,y_1p])
```

**Resolvendo as EDOs e Plotando os gráficos**

```python
t = a
while(t<b):
    X.append(r[0])
    Y.append(r[2])
    r = RK4(r,t) 
    t+=h


plt.xlabel("x [m]")
plt.ylabel("y [m]")
plt.plot(L, 0, 'bo')
plt.plot(0, L, 'bo')
plt.plot(L, 2*L, 'bo')
plt.plot(X, Y, label='P', color = 'g')
plt.plot(X[0], Y[0],'ro')
plt.legend()
plt.xlim(-0.2,2.2*L)
plt.ylim(-0.2,2.2*L)
plt.show()

plt.ylabel("x [m]")
plt.plot(X, label='x')
plt.show()
plt.ylabel("y [m]")
plt.plot(Y, label='x')
plt.show()
```

Este tipo de curva é chamada de "curvas de Lissajous", ela é um caso especial do sistema de três molas descrito acima. Estas curvas ocorrem justamente pois a partícula inicia o seu movimento em uma região em que ela sofre praticamente a mesma influência de todas as molas.
